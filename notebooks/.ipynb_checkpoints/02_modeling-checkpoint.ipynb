# Telco Customer Churn - Predictive Modeling & Evaluation

**Objective:**
Translate the business insights discovered in Phase 1 into a robust, production-ready Machine Learning pipeline. This notebook focuses on strict data leakage prevention, handling class imbalance, optimizing decision thresholds for business value, and conducting deep error analysis.

## 1. Introduction

In the previous notebook (`01_EDA.ipynb`), we identified key churn drivers such as short-term contracts, lack of technical support, and high monthly charges. In this phase, we will build a predictive model to identify customers at high risk of churning, allowing the business to take proactive retention measures.

## 2. Problem Framing

### 2.1 Business Objective
Predict which existing customers are at high risk of churning in the near future so the retention team can proactively target them with personalized campaigns, discounts, or service bundles.

### 2.2 Machine Learning Objective
Build a binary classification model to estimate the probability of churn for each customer. 

### 2.3 Evaluation Metric
- **Primary Metric: Recall (for Churn = 1)**
- **Secondary Metric: ROC-AUC**

**Rationale:** In the telecom industry, a **False Negative** (missing a customer who is actually going to churn) is much more costly than a **False Positive** (giving a retention discount to a customer who was going to stay anyway). Therefore, our model must prioritize maximizing **Recall** to capture as many potential churners as possible, while using ROC-AUC to measure the model's overall ability to distinguish between the two classes.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3

# Scikit-Learn modules
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve

# Machine Learning Models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Environment settings
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid")
```

## 3. Data Loading and Validation

```python
# Define the path to SQLite database
db_path = "../data/database/telco_customer_churn.db"

# Connect to the SQLite database
conn = sqlite3.connect(db_path)

# Load the data into a pandas DataFrame
query = "SELECT * FROM cleaned_churn"
df = pd.read_sql_query(query, conn)


# Close the database connection
conn.close()

print("Dataset Shape:\n", df.shape)
print("Dataset Info:\n", df.info())
display(df.head())
```

```python
# Sanity check: Ensure target variable is properly formatted
if set(df['Churn'].unique()) == {'Yes', 'No'}:
    df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

## 4. Train-Test Split (Preventing Data Leakage)
To simulate a real-world production environment, we must isolate our test set immediately. All feature engineering, imputation, and scaling will be fitted **strictly on the training data** to prevent data leakage.

```python
# Separate features and target  , 'gender', 'PhoneService', 'MultipleLines'
cols_to_drop = ['Churn', 'customerID', 'gender', 'PhoneService', 'MultipleLines']
X = df.drop(columns=cols_to_drop)
y = df['Churn']

# Stratified split to maintain the churn ratio
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    stratify=y, 
    random_state=42
)

print(f"Training set: {X_train.shape}")
print(f"Testing set: {X_test.shape}")
```

## 5. Data Understanding (Training Set)
A quick verification of the training set distributions to inform our preprocessing steps.

```python
train_eda = X_train.copy()
train_eda['Churn'] = y_train

# Target Distribution
print("Churn Rate (%):")
print((y_train.value_counts(normalize=True) * 100).round(2))

# Numeric vs Target (quick signal check)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.boxplot(data=train_eda, x='Churn', y='tenure', ax=axes[0])
axes[0].set_title('Tenure vs Churn')

sns.boxplot(data=train_eda, x='Churn', y='MonthlyCharges', ax=axes[1])
axes[1].set_title('MonthlyCharges vs Churn')

plt.tight_layout()
plt.show()

# Key categorical features (quick check)
sns.countplot(data=train_eda, x='Contract', hue='Churn')
plt.title('Contract vs Churn')
plt.xticks(rotation=20)
plt.show()
```

### Observations

- The dataset shows moderate class imbalance (~26% churn), so Recall is prioritized.

- `tenure` has a clear negative relationship with churn:
  → New customers are more likely to leave.

- `MonthlyCharges` is higher for churn customers:
  → Indicates potential price sensitivity.

- `Contract` type shows strong separation:
  → Expected to be a key predictive feature.

## 6. Feature Engineering
We encapsulate our feature engineering into structured functions to maintain a clear boundary between raw and processed data, ensuring reproducibility.



```python
# Create AvgCharges
X_train['AvgCharges'] = X_train['TotalCharges'] / X_train['tenure'].replace(0, 1)
X_test['AvgCharges'] = X_test['TotalCharges'] / X_test['tenure'].replace(0, 1)

# Feature Categorization
numeric_features = ['tenure', 'MonthlyCharges', 'TotalCharges', 'AvgCharges']
categorical_features = [col for col in X_train.columns if col not in numeric_features]
```

## 7. Data Preprocessing Pipeline
Using Scikit-Learn's `ColumnTransformer` to handle scaling and encoding. This guarantees that parameters (like the mean/std for scaling) learned from the training data are applied consistently to the test data.

```python
# Define transformations
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))
])

# Combine into a single preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

## 8. Model Development

### 8.1 Model Selection Rationale
We will experiment with three distinct algorithms:
1. **Logistic Regression:** Serves as our robust baseline. It is highly interpretable, making it easy to extract feature coefficients to explain predictions to stakeholders.
2. **Random Forest:** A powerful bagging ensemble model that handles non-linear relationships well and is naturally resistant to overfitting.
3. **XGBoost:** A state-of-the-art gradient boosting model known for high performance and execution speed. It includes built-in parameters (`scale_pos_weight`) to handle imbalanced datasets natively.


```python
# Calculate class imbalance ratio 
neg_count = (y_train == 0).sum()  # Number of samples in class 0 (No Churn)
pos_count = (y_train == 1).sum()  # Number of samples in class 1 (Churn)
scale_ratio = neg_count / pos_count

print(f"Class 0 (No Churn): {neg_count}")
print(f"Class 1 (Churn): {pos_count}")
print(f"Imbalance Ratio: {scale_ratio:.2f}")
print(f"→ Using scale_ratio = {scale_ratio:.2f} as class weight for the minority class")
```

### 8.2 Model Training

```python
# Initialize models
models = {
    'Logistic Regression': LogisticRegression(
        max_iter=1000, class_weight='balanced', random_state=42
    ),
    'Random Forest': RandomForestClassifier(
        n_estimators=100, class_weight='balanced', random_state=42  
    ),
    'XGBoost': XGBClassifier(
        use_label_encoder=False, eval_metric='logloss',
        scale_pos_weight=scale_ratio, random_state=42
    )
}

fitted_pipelines = {}

# Train models using the pipeline
for name, model in models.items():
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])
    pipeline.fit(X_train, y_train)
    fitted_pipelines[name] = pipeline
    print(f"Successfully trained {name}")
```

## 9. Model Evaluation (Test Set)
In churn prediction, **Recall** is the most critical evaluation metric.  

Formula:  
$$
Recall = \frac{TP}{TP + FN}
$$


A **False Negative (FN)** occurs when the model predicts that a customer will not churn, but in reality, they do churn.  

This is particularly costly for the business, as it means missing the opportunity to take proactive retention actions, leading to potential revenue loss.

```python
fig, ax = plt.subplots(figsize=(8, 6))

for name, pipeline in fitted_pipelines.items():
    y_pred = pipeline.predict(X_test)
    y_proba = pipeline.predict_proba(X_test)[:, 1]
    
    print(f"\n--- {name} ---")
    print(classification_report(y_test, y_pred))
    
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc_score = roc_auc_score(y_test, y_proba)
    ax.plot(fpr, tpr, label=f"{name} (AUC = {auc_score:.3f})")

ax.plot([0, 1], [0, 1], 'k--', label='Random Guess')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('ROC Curves Performance')
ax.legend(loc='lower right')
plt.show()
```

### Performance Summary

Based on the evaluation metrics on the unseen test set, we observe the following performance across our three models after applying `class_weight='balanced'` (Logistic Regression, Random Forest) and `scale_pos_weight` (XGBoost):

| Model | Accuracy | Class 1 Recall (Churn) | Class 1 Precision | ROC-AUC |
| :--- | :---: | :---: | :---: | :---: |
| **Logistic Regression** | **0.74** | **0.79** | **0.51** | **0.848** |
| Random Forest | 0.78 | 0.49 | 0.60 | 0.837 |
| XGBoost | 0.76 | 0.69 | 0.54 | 0.833 |

### Key Insights & Business Implications

1. **The Power of a Strong Baseline with Class Weights:** By applying `class_weight='balanced'`, **Logistic Regression** significantly outperforms the more complex ensemble tree models across almost all key metrics, achieving the highest Recall (0.79) and ROC-AUC (0.848). This confirms that the relationships between features and churn are largely linear, and that Logistic Regression with proper class balancing is an excellent choice for this dataset.
2. **Class Imbalance Addressed at Training Level:** The use of class weights has substantially narrowed the performance gap between Class 0 and Class 1. However, class imbalance effects are still visible (Class 0 F1-score ~ 0.80-0.86 vs Class 1 F1-score ~ 0.54-0.62).
3. **Further Improvement Possible:** The best model (Logistic Regression) already achieves a strong Recall of **0.79** at the default threshold of 0.5, meaning the model now catches 79% of churners — a significant improvement from the 56% recall without class weights. However, the retention team would still miss about 21% of at-risk customers.




### Bridging to the Next Step
While class weights have significantly improved Recall from the training side, the model still misses about 21% of at-risk customers at the default threshold. To push Recall even higher and minimize lost revenue, we will proceed to **Section 10: Threshold Optimization** to fine-tune the decision boundary, trading a small amount of Precision for even higher Recall.
## 10. Threshold Optimization
By default, models use a threshold of 0.5 to decide between class 0 and 1. However, since missing a churner is expensive, we can deliberately lower the decision threshold to increase our sensitivity (Recall), accepting a slight drop in Precision.

```python
# Let's optimize the threshold using Logistic Regression as our primary model
best_model = fitted_pipelines['Logistic Regression']
y_proba_xgb = best_model.predict_proba(X_test)[:, 1]

# Custom threshold
custom_threshold = 0.40
y_pred_custom = (y_proba_xgb >= custom_threshold).astype(int)

print("Classification Report with Default Threshold (0.50):")
print(classification_report(y_test, best_model.predict(X_test)))

print(f"Classification Report with Optimized Threshold ({custom_threshold}):")
print(classification_report(y_test, y_pred_custom))
```

```python
# Trade-off curves showing how accuracy, precision, and recall change with the decision threshold
from sklearn.metrics import accuracy_score, precision_score, recall_score

y_proba = best_model.predict_proba(X_test)[:, 1]
thresholds = np.linspace(0, 1, 101)

metrics = {'threshold': [], 'accuracy': [], 'precision': [], 'recall': []}
for t in thresholds:
    y_pred = (y_proba >= t).astype(int)
    metrics['threshold'].append(t)
    metrics['accuracy'].append(accuracy_score(y_test, y_pred))
    metrics['precision'].append(precision_score(y_test, y_pred, zero_division=0))
    metrics['recall'].append(recall_score(y_test, y_pred))

metrics_df = pd.DataFrame(metrics)

plt.figure(figsize=(10, 6))
plt.plot(metrics_df['threshold'], metrics_df['accuracy'], label='Accuracy')
plt.plot(metrics_df['threshold'], metrics_df['precision'], label='Precision')
plt.plot(metrics_df['threshold'], metrics_df['recall'], label='Recall')
plt.axvline(0.5, color='k', linestyle='--', alpha=0.6, label='Default (0.5)')
plt.axvline(0.4, color='gray', linestyle=':', alpha=0.6, label='Chosen (0.4)')
plt.xlabel('Threshold')
plt.ylabel('Metric value')
plt.title('Metric Trade-off vs Decision Threshold')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

### Threshold Optimization Insight

As the decision threshold decreases, Recall increases while Precision decreases, highlighting the trade-off between capturing more churners and introducing more false positives.

At the default threshold (0.5), the model misses a significant portion of churn customers (low Recall), which is not acceptable from a business perspective.

By lowering the threshold to 0.35:
- Recall improves significantly (more churners are detected)
- Precision decreases slightly (more false alarms are introduced)

Given that missing a churner is more costly than targeting a non-churner, the threshold of 0.35 provides a better balance aligned with business objectives.

## 11. Model Interpretation
Extracting feature coefficients from our best-performing model (Logistic Regression) to understand what drives its decision-making. For linear models, the absolute value of each coefficient indicates how strongly that feature influences the prediction.

```python
lr_classifier = best_model.named_steps['classifier']
categorical_names = best_model.named_steps['preprocessor'].named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_features)
all_feature_names = numeric_features + list(categorical_names)
# For Logistic Regression, use absolute coefficient values as feature importance
importances = np.abs(lr_classifier.coef_[0])
feature_imp_df = pd.DataFrame({'Feature': all_feature_names, 'Importance': importances})
feature_imp_df = feature_imp_df.sort_values(by='Importance', ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(data=feature_imp_df, x='Importance', y='Feature', palette='magma')
plt.title('Top 10 Feature Importances (Logistic Regression)')
plt.show()
```

## 12. Error Analysis
To truly evaluate our model's readiness for production, we must analyze where it fails. We will compare **False Negatives (FN)** (model predicted Stay, but they Churned) against **True Positives (TP)** (model correctly predicted Churn).

```python
# Create a dataframe for error analysis using the default threshold predictions for baseline comparison
analysis_df = X_test.copy()
analysis_df['Actual_Churn'] = y_test
analysis_df['Predicted_Churn'] = best_model.predict(X_test)

# Extract False Negatives and True Positives
fn_df = analysis_df[(analysis_df['Actual_Churn'] == 1) & (analysis_df['Predicted_Churn'] == 0)]
tp_df = analysis_df[(analysis_df['Actual_Churn'] == 1) & (analysis_df['Predicted_Churn'] == 1)]

print(f"Total False Negatives (Missed Churners): {len(fn_df)}")
print(f"Total True Positives (Caught Churners): {len(tp_df)}")

# Let's compare the characteristics of these two groups
print("\n--- Descriptive Stats for False Negatives (Missed) ---")
display(fn_df[['tenure', 'MonthlyCharges']].describe().T)

print("\n--- Descriptive Stats for True Positives (Caught) ---")
display(tp_df[['tenure', 'MonthlyCharges']].describe().T)
```

**Error Pattern Detection:**
*(Note: Observations based on the stats above)*
- **Tenure:** False Negatives often have a significantly higher average `tenure` compared to True Positives. The model seems to assume that customers who have been with the company for a long time are safe, causing it to miss long-term customers who suddenly decide to leave.
- **Actionable Takeaway:** We might need to engineer a new feature capturing recent behavioral changes (if temporal data were available), or adjust class weights specifically for high-tenure brackets.
